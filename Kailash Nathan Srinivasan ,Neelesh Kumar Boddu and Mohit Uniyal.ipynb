```python
import matplotlib.pyplot as plt
```

```python
def ressample(n):
    x=np.arange(1,101)# np array for 1-100 in order
    p=[]
    b=[]
    c=[] 
    for i in x:
        p.append(1.0/i)# prob of each element getting picked with prob 1/i
    for i in range(0,n):
        y=np.random.choice(x, 1,p)# picking one item from x with prob p
        b.append(y)
    return(b)
```

```python
 def plots(b):
    unique, counts = np.unique(b, return_counts=True)
    d=dict(zip(unique, counts))
    lists = sorted(d.items()) # sorted by key, return a list of tuples

    xx, yy = zip(*lists) # unpack a list of pairs into two tuples

    plt.plot(xx, yy)
    plt.xlabel("VALUE SAMPLED")
    plt.ylabel("COUNT")
    plt.show()
       
```

## Single Run

```python
ressample(1)
```

## Repeating the Algorithm 1000 times 

```python
b=ressample(1000)
```

```python
plots(b)
```
